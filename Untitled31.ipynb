```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([ # hours studied, sleep, attendance
    [2,6,65],
    [3,7,70],
    [4,6,75],
    [5,7,80],
    [6,6,82],
    [7,7,85],
    [8,8,90],
    [9,7,92],
    [10,8,95],
    [11,7,97]
])

y= np.array([
    [50],
    [55],
    [60],
    [65],
    [70],
    [75],
    [82],
    [86],
    [90],
    [94]
])


w=np.random.randn(3,1)*0.01
b=0

learning_rate=0.01
iterations=1000

x_scaled=x/np.max(x,axis=0)

losses = []
for i in range(iterations):

  y_pred=x_scaled@w+b
  error=y_pred-y

  grad_w=(x_scaled.T@error)/x.shape[0]
  grad_b=np.mean(error)

  w=w-learning_rate*grad_w
  b=b-learning_rate*grad_b

  if i%100==0:
    loss=np.mean(error**2)
    losses.append(loss) # Append the calculated loss to the list
    print(f"iteration{i}:,loss={loss}")

hours=float(input("enter study hours: "))
sleep=float(input("enter sleep hours: "))
attendance=float(input("enter attendance percentage:"))

new_student=np.array([[hours,sleep,attendance]])
new_student_scaled=new_student/np.max(x,axis=0)
prediction=new_student_scaled@w+b
print("predicted score:", prediction [0][0])


plt.plot(losses)
plt.xlabel("iterations")
plt.ylabel("loss")
plt.title("training loss curve")
plt.show()
```

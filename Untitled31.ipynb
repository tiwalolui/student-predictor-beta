```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([ # hours studied, sleep, attendance
    [2,6,65],
    [3,7,70],
    [4,6,75],
    [5,7,80],
    [6,6,82],
    [7,7,85],
    [8,8,90],
    [9,7,92],
    [10,8,95],
    [11,7,97]
])

y= np.array([
    [50],
    [55],
    [60],
    [65],
    [70],
    [75],
    [82],
    [86],
    [90],
    [94]
])


w=np.random.randn(3,1)*0.01
b=0

learning_rate=0.01
iterations=1000

x_scaled=x/np.max(x,axis=0)

for i in range(iterations):

  y_pred=x_scaled@w+b
  error=y_pred-y

  grad_w=(x_scaled.T@error)/x.shape[0]
  grad_b=np.mean(error)

  w=w-learning_rate*grad_w
  b=b-learning_rate*grad_b

  if i%100==0:
    loss=np.mean(error**2)
    print(f"iteration{i}:,loss={loss}")


new_student=np.array([[10,10,95]])
new_student_scaled=new_student/np.max(x,axis=0)
predicted_score=new_student_scaled@w+b
print("predicted score:",predicted_score)
```
